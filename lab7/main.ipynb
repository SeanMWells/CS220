```python
import csv

fifa_file = open('Fifa20.csv', encoding='utf-8')
file_reader = csv.reader(fifa_file)
player_data = list(file_reader)
fifa_file.close()
header = player_data[0]
player_data = player_data[1:]
for row in player_data:
    for idx in [2,3,4,7,8,12]:
        row[idx] = int(row[idx])
```

```python
def highest(catagory1=None, catagory2=None):
    highest = None
    high_index = 0
    for row in range(len(player_data)):
        curr = player_data[row][header.index(catagory1)]
        if highest == None or curr > highest:
            highest = curr
            high_index = row
    return player_data[high_index][header.index(catagory2)]
```

```python
def get_column(col_name, start=1, end=None):
    L = []
    for row in range(len(player_data[start:end])):
        L.append(player_data[row][header.index(col_name)])
    return L
```

```python
#q5
get_column("Nationality", 1, 6)
```

```python
def get_sorted_column(col_name, start=1, end=None):
    L = []
    for row in range(len(player_data[start:end])):
        L.append(player_data[row][header.index(col_name)])
        L.sort()
    return L
```

```python
#q6
get_sorted_column("Name")[0:5]
```

```python
def average(catagory):
    total = 0
    for row in range(len(player_data)):
        total += player_data[row][header.index(catagory)]
    return total / (len(player_data))
```

```python
#q7
round(average("Value"), 2)
```

```python
#q8
round(average("Height(cm)"), 2)
```

```python
def position(position):
    L = []
    for row in range(len(player_data)):
        curr_position = player_data[row][header.index("Player_Position")]
        if curr_position == position:
            L.append(player_data[row][header.index("Name")])
    return len(L)
```

```python
#q9
position("CAM")
```

```python
def player_per(catagory):
    Dict = {}
    for row in range(len(player_data)):
        curr_catagory = player_data[row][header.index(catagory)]
        if curr_catagory not in Dict:
            Dict[curr_catagory] = 1
        else:
            Dict[curr_catagory] += 1
    return Dict
```

```python
#q10
player_per("Nationality")
```

```python
def most_players():
    most = None
    country = None
    Dict = player_per("Nationality")
    for row in range(len(Dict)):
        curr_num = Dict[player_data[row][header.index("Nationality")]]
        if most == None or curr_num > most:
            most = Dict[player_data[row][header.index("Nationality")]]
            country = player_data[row][header.index("Nationality")]
    return country
```

```python
def player_to_dict(player_id):
    index = index_finder(player_id)
    Dict = {}
    for each in range(len(player_data[index])):
        Dict[header[each]] = player_data[index][each]
    return Dict
        

def index_finder(player_id):
    for row in range(len(player_data)):
        curr_ID = player_data[row][header.index("ID")]
        if curr_ID == player_id:
            return row
```

```python
def average_club(catagory, club):
    total = 0
    L = []
    for row in range(len(player_data)):
        if player_data[row][header.index("Club")] == club:
            total += player_data[row][header.index(catagory)]
            L.append(player_data[row][header.index(catagory)])
    return total / len(L)
```

```python
#q19
round(average_club("Stamina", "FC Bayern München"), 2)
```

```python
def wage_per_club():
    L = list(set(get_column("Club")))
    Dict = {}
    for row in range(len(L)):
        if L[row] not in Dict:
            Dict[L[row]] = round(average_club("Wage", L[row]), 2)
    return Dict
```
